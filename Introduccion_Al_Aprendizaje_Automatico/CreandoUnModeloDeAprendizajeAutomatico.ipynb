```python
import pandas
!python -m wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/graphing.py
!python -m wget https://raw.githubusercontent.com/MicrosoftDocs/mslearn-introduction-to-machine-learning/main/Data/doggy-boot-harness.csv

##hacer un diccionario para tamaños de arranque 
##y tamaño de arnes 

data = {
    'boot_size' : [ 39, 38, 37, 39, 38, 35, 37, 36, 35, 40, 
                    40, 36, 38, 39, 42, 42, 36, 36, 35, 41, 
                    42, 38, 37, 35, 40, 36, 35, 39, 41, 37, 
                    35, 41, 39, 41, 42, 42, 36, 37, 37, 39,
                    42, 35, 36, 41, 41, 41, 39, 39, 35, 39
 ],
    'harness_size': [ 58, 58, 52, 58, 57, 52, 55, 53, 49, 54,
                59, 56, 53, 58, 57, 58, 56, 51, 50, 59,
                59, 59, 55, 50, 55, 52, 53, 54, 61, 56,
                55, 60, 57, 56, 61, 58, 53, 57, 57, 55,
                60, 51, 52, 56, 55, 57, 58, 57, 51, 59
                ]
}


##se convierte en una tabla de panda

dataset = pandas.DataFrame(data)

##se imprime el data
dataset

##se importa la biblioteca para crear el modelo

import statsmodels.formula.api as SMF

##primero se define  la formula usando una sintaxis especial
formula = "boot_size ~ harness_size"

##se crea el modelo, pero aun no se entrena
model = SMF.ols(formula = formula, data = dataset)

##se crea el primero modelo 
##aun falta tener parametros internos

if not hasattr(model,'params'):
    print('El modelo seleccioonado aun no tiene parametros, se nececita que entrenar')
    

##se importa la libreria que va hacer el trabajo para nosotros
import graphing

##se entrena el modelo para cree una linea que se ajusta a nuestro datos, este metodo realiza el trabajo duro,
fitted_model = model.fit()

# Print information about our model now it has been fit
print("Se encontro los siguientes parametros en el modelo:\n" +
        f"Pendiente de linea: {fitted_model.params[1]}\n"+
        f"Intercepcion de lineas: {fitted_model.params[0]}")

#muestra grafico del resultado
graphing.scatter_2D(dataset,    label_x="harness_size", 
                                label_y="boot_size",
                                trendline=lambda x: fitted_model.params[1] * x + fitted_model.params[0]
                                )

##harnes indica el tamaño  de arnes que nos interesa
harness_size = {'harness_size':[52.2]}
##se usa el modelo para predecir el tamaño  de botas que le quedara al perro
approximate_boot_size = fitted_model.predict(harness_size)

#se imprime el resultado

print('Estimado  del tamaño de la bota')
print(approximate_boot_size[0])
```
